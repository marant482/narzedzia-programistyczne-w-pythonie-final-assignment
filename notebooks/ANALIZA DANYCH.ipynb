### Łączenie zbiorów

Stwórzmy jeden złączony Dataframe z tych trzech zbiorów.

```python
wszystkie_dane=pd.merge(pozary, powierzchnie[['TERYT', 'Powierzchnia [ha]']], on='TERYT', how='left')
```

```python
wszystkie_dane=pd.merge(wszystkie_dane, populacja[['TERYT', 'Ludność']], on='TERYT', how='left')
```

```python
wszystkie_dane
```

Nie będziemy już potrzebować kolumny TERYT i Powiat, więc usuńmy je dla klarowności

```python
wszystkie_dane=wszystkie_dane.drop(['TERYT', 'Powiat'], axis=1)
wszystkie_dane
```

przygotujmy jeszcze zagregowane dane na poziomie Województw do analizy z danymi alkohol

```python
wszystkie_dane_wojewodztwo= wszystkie_dane.groupby('Województwo').agg({
    'Liczba Pożarów': 'sum',
    'Powierzchnia [ha]': 'sum',
    'Ludność': 'sum'
}).reset_index()
wszystkie_dane_wojewodztwo
```

dopasujmy nazwy w zbiorze alkohole_wojewodztwo

```python
alkohol_wojewodztwo=ppr.usun_woj(alkohol_wojewodztwo)
alkohol_wojewodztwo=ppr.litery_na_male(alkohol_wojewodztwo)
alkohol_wojewodztwo
```

dopasujmy też gminy miejskie, które są w alkohol_miejscowosc

```python
wszystkie_dane_miejscowosc=wszystkie_dane[wszystkie_dane["Gmina"].isin(alkohol_miejscowosc["Miejscowość"])]
wszystkie_dane_miejscowosc
```

```python
wszystkie_dane_miejscowosc = wszystkie_dane_miejscowosc.groupby('Gmina').agg({
    'Liczba Pożarów': 'sum',
    'Powierzchnia [ha]': 'sum',
    'Ludność': 'sum'
}).reset_index()
```

```python
wszystkie_dane_miejscowosc
```
